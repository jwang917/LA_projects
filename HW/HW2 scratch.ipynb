```python
#Write out the forward elimination algorithm function:
def REF_basic(B):
    #get the coefficient matrix Bcof from augmented matrix B.
    Bcof = B[:,:-1]
    print(Bcof)
    #check invertibility of Bcof by compare #row and #columns, and see if the rank is full.
    if (np.shape(Bcof)[0] != np.shape(Bcof)[1]) or (np.linalg.matrix_rank(Bcof) != np.shape(Bcof)[0]):
        print("Matrix B is not invertible. Bad input!")
        return 
    else:
        #use a for loop to iterate all the rows
        nrows = np.shape(B)[0]
        for i in range(nrows):
            #Use a variable sign to siginal the forward elimination process is done.
            sign = True
            while(sign):
                #check the zero in 1st element in i.
                pivot = B[i,i]
                if(pivot != 0):
                    #print("The pivot now is",i,"in the",i,"entry of diagonal.")
                    if i == (nrows-1):
                        print("we reach the last pivot, time to go home!")
                        break
                    #doing elimination.
                    print("start forward elimination for",i+1,"th column")
                    for ii in range(i+1,nrows):
                        B[ii] = (-B[ii,i]/B[i,i])*B[i]+B[ii]
                        #print("The row below pivot becomes:",B[ii])
                    #print("For pivot in",i,"column, the elimination yields:",B)
                    sign=False
                else:
                    #find non-zero entry below and do swap if it is possible.
                    for ii in range(i+1,nrows): 
                        #start one row below ith row, move up all the non-zero entries.
                        if B[ii,i]!=0:
                            
                            #swap row ii with i.
                            container = np.copy(B[ii-1]) #use container to store row i
                            B[ii-1] = B[ii]     #replace row i with new row
                            B[ii] = container# put container into new row
                            
                            #print("Finish swap row",ii,'with',ii-1,"in the iteration:",i)
                            break
                            #We only need one swap.
                        #if row ii reach the end ie ii=nrows-1, print error message. ie:"the ith column is zero column."
                        else:
                            if(ii==nrows-1):
                                print("Error: the",i,"th column is zero column.")
                                break
                            else:
                                continue
    return(B)
```

```python
def REF(B):
    #Given any size of matrix, conduct the forward elimination.
    #use a for loop to iterate all the rows
    nrows = np.shape(B)[0]
    for i in range(nrows):
        #Use a variable sign to siginal the forward elimination process is done.
        sign = True
        while(sign):
            #check the zero in 1st element in i.
            pivot = B[i,i]
            if(B[i,i]!=0):
                print("The pivot now is",i,"in the",i,"entry of diagonal.")
                if i == (nrows-1):
                    print("we reach the last pivot, time to go home!")
                    print("The B is this",B)
                    break
                #doing elimination.
                for ii in range(i+1,nrows):
                    B[ii] = (-B[ii,i]/B[i,i])*B[i]+B[ii]
                    print("The row below pivot becomes:",B[ii])
                print("For pivot in",i,"column, the elimination yields:",B)
                sign=False
            else:
                #find non-zero entry below and do swap if it is possible.
                for ii in range(i+1,nrows): 
                    #start one row below ith row, move up all the non-zero entries.
                    if B[ii,i]!=0:
                        print("We find a pivot candidate!!!,it is the",ii,"th row")
                        #swap row ii with i.
                        container = np.copy(B[ii-1]) #use container to store row i
                        B[ii-1] = B[ii]     #replace row i with new row
                        B[ii] = container# put container into new row
                        print("Container is :",container)
                        print("Finish swap row",ii,'with',ii-1,"in the iteration:",i)
                        break
                        #clear out container
                        #We only need one swap.
                    #if row ii reach the end ie ii=nrows-1, print error message. ie:"the ith column is zero column."
                    else:
                        if(ii==nrows-1):
                            print("Error: the",i,"th column is zero column.")
                            break
                        else:
                            continue
    return B
```

```python
def chkshape(array):
    #write a new function: chkshape(array). Output -1 means error; 1 it is square; 0 it is rectangle.
    if np.ndim(array) != 2:
        print("Error, the matrix is not 2-Dimension.")
        return -1
    else:
        shape = np.shape(array)
        if (shape[0]-shape[1]) == 0:
            return 1
        else: 
            return 0

        
def checkUpper(array):
    #write a function to check the upper triangle property of any given matrix.
    #Output: True as upper tirangle shape, False as not upper triangle shape.
    #Learned from [https://www.geeksforgeeks.org/program-check-matrix-upper-triangular/]
    #This should be done after confirmation of being square matrix.
    dim = np.shape(array)[0]
    for i in range(1,dim):
        for j in range(0,i):
            if array[i][j]!=0:
                return False
    return True
        
#Compute the unknowns by backward substitution.
#Learned from [https://codereview.stackexchange.com/questions/110039/back-substitution-method-for-solving-linear-system]
def backsub(array,vector):
    if chkshape(array) !=1:
        if chkshape(array) == 0:
            print("Error message: Matrix\n",array," \nis rectangle. ")
            return
    else:
        if checkUpper(array) == False:
            print("Error message: Matrix's shape is not upper-triangle.\n","Please re-enter your matrix")
            return
    ls = np.zeros_like(vector)
    size = np.shape(vector)[0]
    
    #Find the last index's value of vector.
    for i in range(size-1,-1,-1): #index starts from n-1 to 0.
        #Check if all diagonal entries is nonzero. If it is not, then print error and return.
        if array[i,i] == 0:
            print("Error message: the diagonal entries have number 0.")
            return
        if i == size - 1:
            #Start to iterate from the bottom row, get the last entry's value.
            x = (vector[i])/array[-1,-1]
            #Then go up one row. compute the result of x(index) by using what is stored in the result list(ls).
        else:
            #compute the result of x(index).
            x = (vector[i]-np.sum(array[i,i+1:] * ls[i+1:]))/array[i,i]
        #store the value of unknown into ls, with index starting backward. ie: xn,xn-1,...,x2,x1
        ls[i]=x
    print(ls)
 

def MySolve(A,b):
    #do the foward elimination first. A is n*n, b is n*1
    aug = np.column_stack((A,b))
    #Forward elimination result.
    uRes = REF_basic(aug)
    #Then get the coefficient matrix, and constant after forward elimination to do the back substitution.
    cof = aug[:,:-1]
    const = aug[:,-1]
    backsub(cof,const)
    return "Finish solving Ax=b."
```
